```python
from torch.utils.data import Dataset


class QADataset(Dataset):
    def __init__(self, df_train, tokenizer, max_length):
        self.input_ids = []
        self.attn_masks = []
        self.labels = []
        
        dataset = ""
        context=df_train.context.tolist()
        questions=df_train.question.tolist()
        answers=df_train.answer_text.tolist()
        starts = df_train.answer_start.tolist()
        c, q, a, e = "<|startoftext|>[CONTEXT]: ", "\n[QUESTION]: ", "\n[ANSWER]: ", "<|endoftext|>\n"

        for ind in range(len(questions)):
            start = 0
            end = min(300, len(context[ind]))
            stride = 150
            s_pos = starts[ind]
            while end < len(context[ind]):
                temp = c + context[ind][start:end]
                temp += q + questions[ind]
                e_pos = s_pos + len(answers[ind])-1
                if s_pos >=start and e_pos<end:
                    ans = answers[ind]
                elif s_pos >= end or e_pos < start:
                    ans = "NONE"
                    start = end-stride
                    end = min(end - stride +300, len(context[ind]))
                    continue
                else:
                    ans = context[ind][max(s_pos, start):min(e_pos+1, end)]
                temp += a + ans
                temp += e
                start = end-stride
                end = min(end - stride +300, len(context[ind]))
                # tokenize
                encodings_dict = tokenizer(temp, truncation=True,
                                           max_length=max_length, padding="max_length")
                # append to list
                self.input_ids.append(torch.tensor(encodings_dict['input_ids']))
                self.attn_masks.append(torch.tensor(encodings_dict['attention_mask']))
                self.labels.append(ans)

#                 dataset += temp
            temp = c + context[ind][start:end]
            temp += q + questions[ind]
            e_pos = s_pos + len(answers[ind])-1
            if s_pos >=start and e_pos<end:
                ans = answers[ind]
            elif s_pos >= end or e_pos < start:
                continue
            else:
                ans = context[ind][max(s_pos, start):min(e_pos+1, end)]
            temp += a + ans
            temp += e
            # tokenize
            encodings_dict = tokenizer(temp, truncation=True,
                                           max_length=max_length, padding="max_length")
            # append to list
            self.input_ids.append(torch.tensor(encodings_dict['input_ids']))
            self.attn_masks.append(torch.tensor(encodings_dict['attention_mask']))
            self.labels.append(ans)
    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return self.input_ids[idx], self.attn_masks[idx], self.labels[idx]
```

```python
# Data load function
def load_sentiment_dataset(tokenizer, df):

    train_dataset = QADataset(df, tokenizer, max_length=1024)

    # return
    return train_dataset
```
